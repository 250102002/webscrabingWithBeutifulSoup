```python
from bs4 import BeautifulSoup
import requests 
import csv 


link = r"https://egypt.souq.com/eg-en/labtops/s/?as=1&section=2&page="
```

```python
# Lab_name = soup.find_all('div',{'class':'col col-info item-content'})
# Lab_link = soup.find_all('a',{'class':"itemLink sk-clr2 sPrimaryLink"})
with open("LabtopsData.csv","w") as file :
    writer = csv.writer(file)
    writer.writerow(['name','color','discrebtion','price','rate on soaq.com'])
```

```python
for i in range(1,10):
    lk = link+str(i)
    site = requests.get(lk)
    content = site.content
    soup = BeautifulSoup(content,"lxml")
    """data we need"""
    Lab_name = soup.find_all('div',{'class':'col col-info item-content'})
    num_iter = len(Lab_name)
    lnks = []
    for i2 in range(num_iter):
        lnks.append(Lab_name[i2].find('a').attrs['href'])
    for ln in lnks :
        try :
            Lab_site = requests.get(ln)
            Con = Lab_site.content
            Soup=BeautifulSoup(Con,"lxml")
            Lab_colors = Soup.find_all("a",{'class':'value size-value'})
            Lab_discreb=Soup.find_all('ul',{'class':"generatedBullets"})
            Lab_price = Soup.find_all('h3',{'class':'price is sk-clr1'})
            Lab_rate = Soup.find_all('div',{'class':'avg'})
            Lab_na = Soup.find_all('div',{'class':'small-12 columns product-title'})
            with open('LabtopsData.csv',"a+") as file :
                writer = csv.writer(file)
                for i3 in range(1):
                    writer.writerow([Lab_na[i3].find('h1').text.strip(),Lab_colors[i3].text.strip(),Lab_discreb[i3].text.strip(),Lab_price[i3].text.strip(),Lab_rate[i3].text.strip()])
            print("web switched")
        except :
            print('error')
            continue
```
